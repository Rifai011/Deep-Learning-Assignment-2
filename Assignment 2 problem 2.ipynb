```python
import tarfile
import os

# Define the path to the dataset (modify this path as needed)
tar_path = r"C:\Users\Rifai Rahman\Documents\Deep Learning Assignment 2/ILSVRC2010_images_val.tar"  # Update with actual file path
extract_path = r"C:\Users\Rifai Rahman\Documents\Deep Learning Assignment 2"  # Change to where you want to extract

# Extract the dataset
with tarfile.open(tar_path, "r:") as tar:
    tar.extractall(extract_path)

print(f"Extraction complete! Files are in {extract_path}")
```

```python
extract_path= r"C:\Users\Rifai Rahman\Documents\Deep Learning Assignment 2\dataset\val"


# List the first 10 files in the extracted dataset
files = os.listdir(extract_path)
print("First 10 extracted files:", files[:10])
```

```python
from PIL import Image
import matplotlib.pyplot as plt

# Filter out image files (assuming JPG format for ILSVRC2010)
image_files = [f for f in files if f.endswith('.JPEG')]  # Modify if images have a different extension

# Load and display the first image
if image_files:
    img_path = os.path.join(extract_path, image_files[0])
    img = Image.open(img_path)

    plt.imshow(img)
    plt.axis("off")  # Hide axes
    plt.show()
else:
    print("No image files found.")
```

```python
fig, axes = plt.subplots(2, 3, figsize=(12, 8))  # 2 rows, 3 columns

for ax, img_file in zip(axes.ravel(), image_files[:6]):  # Display first 6 images
    img = Image.open(os.path.join(extract_path, img_file))
    ax.imshow(img)
    ax.axis("off")

plt.show()
```
